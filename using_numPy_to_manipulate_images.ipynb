```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
```

## The Base Image

```python
# This is the base image provided by scipy and plotted on a graph with matplotlib.
img = misc.face()
plt.imshow(img)
```

## Gray Image

```python
# Turning the image gray by converting the pixels to an RGB value. Then mulitply the RGB image by a
# gray value scale to get the final gray image.
img = misc.face()
img_rgb = img / 255
gray_vals = np.array([0.2126, 0.7152, 0.0722])
gray_img = img_rgb @ gray_vals
plt.imshow(gray_img, cmap='gray')
```

## Solarization of Image

```python
# Solarizing the image by reversing the pixel values
img = misc.face()
solar_img = 255- img
plt.imshow(solar_img)
```
